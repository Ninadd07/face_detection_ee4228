```python
import numpy as np

embs = np.load("embeddings/jessica/Jessica_Dark_embeddings.npy")  # shape: (N, 512)
print(embs.shape)
```

```python
def cosine_similarity(a, b):
    a = a / np.linalg.norm(a)
    b = b / np.linalg.norm(b)
    return float(np.dot(a, b))

i, j = 0, 10  # pick two indices
sim = cosine_similarity(embs[i], embs[j])
print(sim)
```

```python
for i in range(0, min(10, len(embs) - 1)):
    sim = cosine_similarity(embs[0], embs[i])
    print(i, sim)
```
